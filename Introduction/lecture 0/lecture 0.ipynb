### 一,任务分类  

#### 学习机器学习,就是为了简化生活中的任务.不同方法适用于不同任务,那么贴合业务逻辑的业务,个人认为可以分类如下:  

1)预测类:预测连续型数值,输出为实数.比如股价预测,销量预测等  

2)分类:将样本划分到离散的类别中,输出为类别标签/类别概率.比如:文本情感,商品属性等  

3)聚类: 在无标签数据中，根据样本的相似性自动划分成若干簇，挖掘数据的内在分布规律，属于无监督学习.比如用户分群,客户画像聚类  

#### 以上三类为常见的任务分类.  

4)生成类:根据已有的数据,生成新的样本.比如:图像生成,文本生成等  

5)优化类:在约束条件下,找到目标函数的最优解,输出为决策变量的最优取值,比如强化学习,规划等   
 
6)表征类:将数据映射到一个低维空间中,为后续建模做基础.比如文本嵌入,图像特征提取
